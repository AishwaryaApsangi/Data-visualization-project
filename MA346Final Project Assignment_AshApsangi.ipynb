## CHANGING DATA TYPES:

> I have defined a function to convert specified columns in df to a numeric format, addressing common formatting issues like commas, percentages, and textual representations of large numbers. The columns  'Population (Millions)', 'GDP (Billions, PPP)', 
'GDP per Capita (PPP)', 'FDI Inflow (Millions)', and 'Unemployment (%)' are targeted for this conversion. These columns need to be converted to floats, a data type compatible with missing values, ensuring they can still be used in calculations. This careful data cleaning helps preserve the dataset's completeness, enabling a thorough analysis.

How can the columns be accurately processed and converted in a consistent numeric format suitable for analysis? 

```python
def convert_column_to_float(column):
    """
    Converts a dataframe column to float type after cleaning non-numeric characters.

    Returns:
    None: The dataframe is modified in place, and the columns are converted to float type.

    """
    df[column] = df[column].replace({'ppl\.': '', '\$': '', ',': '', ' million': 'e6', ' billion': 'e9', '%': '', '-': ''}, regex=True) # removes non-numeric characters such as commas, spaces, 'Millions',  'Billions', and '%' from a dataframe column. 
    df[column] = pd.to_numeric(df[column], errors='coerce').astype(float) # Converts to float type




# Apply the conversion to each column
for column in [ 'Population (Millions)',  'GDP (Billions, PPP)',  'GDP per Capita (PPP)',  'Unemployment (%)',   'FDI Inflow (Millions)']:
    convert_column_to_float(column)
    
# Check the dtypes to confirm the changes
df.dtypes
```

> The specified columns in df have been successfully converted to numeric (float) formats, with non-numeric characters and formatting removed. The DataFrame is now ready for quantitative analysis with these standardized numeric columns. I also converted the 'GDP per Capita (PPP)' and  'GDP (Billions, PPP)' columns in df to a numeric (float) format by removing currency symbols and then transforming the column to a numeric data type, thereby addressing any parsing errors. The updated data types of all columns are also verified to confirm this transformation.

# 2. MERGING DF AND DF2:

> Next, the script below cleans and aligns the country name columns in both df and df2 to ensure consistency. The motivation for the code before is so that an inner join is performed to merge these datasets on the common country names, creating a new DataFrame merged_df. The redundant 'Country or region' column is then removed, and 'Country Name' is renamed to 'Country' for clarity.

So how can we ensure that the datasets are aligned by country names, allowing for a precise and accurate inner join merge?

```python

df = df.copy()  # Create a copy of the first dataframe to avoid modifying it directly

df['Country Name'] = df['Country Name'].str.strip() # Remove leading and trailing whitespace from the 'Country Name' column

df2 = df2.copy() # Create a copy of the second dataframe to avoid modifying it directly
df2['Country or region'] = df2['Country or region'].str.strip() # Remove leading and trailing whitespace from the 'Country or region' column to normalize the data


merged_df = pd.merge(df, df2, left_on='Country Name', right_on='Country or region', how='inner') # Perform the inner join on the common key
merged_df.drop('Country or region', axis=1, inplace=True) # Drop the 'Country or region' column from the merged DataFrame


merged_df.rename(columns={'Country Name': 'Country'}, inplace=True)# Optionally, rename 'Country Name' to 'Country' for clarity
merged_df # Display the resulting DataFrame
```

The merged_df now contains a consolidated dataset with matched and merged data from both df and df2. It holds only the records where the country names are aligned, providing a cohesive view of the combined data.

# 3. HYPOTHESIS TESTING/STATISTICAL ANALYSIS:

> For the statistical analysis part of the project, the script will perform a linear regression analysis (which includes hypothesis testing) between 'GDP per Capita (PPP)' and 'Happiness Score' in the merged_df DataFrame, using the stats.linregress function from the SciPy library. This analysis aims to statistically quantify the relationship between these two variables. 

What is the relationship between GDP per Capita (PPP) and happiness scores, and how statistically significant is this relationship?

```python
from scipy import stats

# Assuming 'merged_df' is the DataFrame, 'GDP per Capita (PPP)' is the GDP per capita column,
# and 'Happiness Score' is the happiness score column.

# Perform linear regression
model = stats.linregress(merged_df['GDP per Capita (PPP)'], merged_df['Happiness Score'])


# Extract the model parameters
beta0 = model.intercept  # Intercept (Î²0)
beta1 = model.slope      # Slope (Î²1)
r_value = model.rvalue   # Correlation coefficient
r_squared = r_value**2   # Coefficient of determination (R-squared)
p_value = model.pvalue   # p-value for hypothesis test


# You can output all the values in a single cell by creating a dictionary:
regression_results = {
    "Intercept (Î²0)": beta0,
    "Slope (Î²1)": beta1,
    "R-squared value": r_squared,
    "p-value": p_value,
    "Significance": ("The correlation between GDP per capita (PPP) and happiness score is statistically significant." if p_value < 0.05 else "There is no significant correlation between GDP per capita (PPP) and happiness score.")
}

regression_results

```

By calculating the slope, intercept, correlation coefficient, and p-value, the script assesses the strength and significance of the correlation, addressing the project's goal of exploring and quantifying relationships between key economic and well-being indicators.  The p-value obtained from the regression analysis tests the null hypothesis and checks whether there is no relationship between GDP per Capita (PPP) and happiness scores against the alternative hypothesis that there is a relationship.  Given the p-value of 9.274864749171573e-23  is below the conventional threshold of 0.05 and that the slope is positive, it can be concluded that the relationship is statistically significant and that there is a positive correlation between GDP per Capita (PPP) and happiness (quantified by happiness scores).
